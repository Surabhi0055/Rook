```python
import pandas as pd

ratings = pd.read_csv("../dataset/ratings.csv")
ratings.head()
ratings.shape
```

```python
user_counts = ratings['user_id']
user_counts.describe()
```

```python
active_users = user_counts[user_counts >= 10].index
ratings = ratings[ratings['user_id'].isin(active_users)]
ratings.shape
```

```python
counts = ratings['book_id'].value_counts()
counts.describe()
```

```python
books = counts[counts >= 10].index
ratings = ratings[ratings['book_id'].isin(books)]
```

```python
print("Users:", ratings['user_id'].nunique())
print("Books:", ratings['book_id'].nunique())
print("Total Ratings:",ratings.shape[0])
```

```python
ratings.groupby('user_id')['rating'].mean().describe()
```

```python
ratings.to_csv("../dataset/ratings_processed.csv", index=False)
```
