```python
import pandas as pd

books = pd.read_csv("../dataset/books_cleaned.csv")
book_tags = pd.read_csv("../dataset/book_tags.csv")
tags = pd.read_csv("../dataset/tags.csv")

books.head()
tags.head()

```

```python
book_tags = book_tags.merge(tags, on="tag_id", how="left")
book_tags.head()
```

```python
book_tags = book_tags[["goodreads_book_id", "tag_name"]]
```

```python
book_tags_grouped = book_tags.groupby("goodreads_book_id")["tag_name"] \
    .apply(lambda x: " ".join(x)) \
    .reset_index()

book_tags_grouped.head()
```

```python
books = books.merge(
    book_tags_grouped,
    left_on="book_id",
    right_on="goodreads_book_id",
    how="left"
)

books.head()
```

```python

books["content_features"] = (
    books["title"] + " " +
    books["authors"] + " " +
    books["tag_name"].fillna("")
)

books[["title", "content_features"]].head()

books.head()
```

```python
books["content_features"] = books["content_features"].fillna("")
books["content_features"] = books["content_features"].astype(str)
print(books["content_features"].isnull().sum())
```

```python
books.to_csv("books_genre.csv", index=False)
```
