```python
import pandas as pd

books = pd.read_csv("../dataset/books.csv")
books.head()
```

```python
books.shape
books.info()
books.isnull().sum()
```

```python
books = books.drop_duplicates(subset="title")
```

```python
#missing values
books = books.dropna(subset=["title"])
books["authors"] = books["authors"].fillna("Unknown")

books["original_publication_year"] = books["original_publication_year"].fillna(0)
books["original_publication_year"] = books["original_publication_year"].astype(int)
books = books[books["original_publication_year"] > 1800]

books["language_code"] = books["language_code"].fillna("unknown")
```

```python
books["title"] = books["title"].str.lower().str.strip()
```

```python
import re
books["title"] = books["title"].apply(
    lambda x: re.sub(r"[^a-zA-Z0-9 ]", "", x)
)
```

```python
books = books[books["ratings_count"] > 50]
books = books[books["work_ratings_count"] > 50]
books = books.drop_duplicates("book_id")

```

```python
books["content"] = (
    books["title"] + " " +
    books["authors"] + " " +
    books["original_publication_year"].astype(str)
)
```

```python
books["image_url"] = books["image_url"].fillna("book_not_found.jpg")
```

```python
books.to_csv("books_cleaned.csv", index=False)
```
